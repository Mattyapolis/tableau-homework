```python
import pandas as pd
import numpy as np
import csv
```

```python
csv = "./data/201812.csv"
dec_2018 = pd.read_csv(csv)


```

```python
dec_2018.head()
```

```python
df=dec_2018.groupby("bikeid").agg({"tripduration" : ["median", "sum", "count"]})
df["month"] = "Dec"
df["year"] = 2018

df



df.to_csv("bikeid_201812.csv",index=True)


# df = df[["year", "month", "bikeid", "tripduration", "start station id"]]
# df = df.rename(columns={"bikeid" : "bike_id", "tripduration" : "median_trip_duration", "start station id" : "ride_count"})
```
